```python
def remove_nested_field(selectors):
    def wrap_(obj):
        def wrap(sels, obj):
            if len(sels) == 1:
                del obj[sels[0]]
            else:
                if sels[0] == list:
                    body = sels[1:]
                    for one in obj:
                        wrap(body, one)
                else:
                    wrap(sels[1:], obj[sels[0]])
        wrap(selectors, obj)
        return obj
    return wrap_


def remove_fields(fields):
    def wrap(args):
        args = deepcopy(args)
        for field in fields:
            if type(field) == list:
                remove_nested_field(field)(args)
            else:
                if field in args:
                    del args[field]

        return args

    return wrap
```

```python
def to_args(content):
    combined = {}
    if 'metadata' in content:
        combined.update(content['metadata'])
    if 'spec' in content:
        combined.update(content['spec'])
    return combined
```

```python
def sort_field_by(selectors, key):
    def wrap_(obj):
        def wrap(sels, obj):
            if len(sels) == 0:
                obj.sort(key=key)
            else:
                if sels[0] == list:
                    for one in obj:
                        wrap([], one)
                else:
                    wrap(sels[1:], obj[sels[0]])
        wrap(selectors, obj)
    return wrap_
```

```python
def remove_empty_list(content):
    if type(content) == dict:
        to_remove = set()
        for key, val in content.items():
            if val == []:
                to_remove.add(key)
        return {k: remove_empty_list(v) for k, v in content.items() if k not in to_remove}
    elif type(content) == list:
        return [remove_empty_list(one) for one in content]
    else:
        return content
```
