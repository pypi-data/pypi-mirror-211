```python
x=np.array(data['age'])
y=np.array(data['charges'])
x=x.reshape(1338 ,1) 
# .reshape()으로 차원 맞추기
y=y.reshape(1338 ,1)
lr = LinearRegression()
lr.fit(x,y)
```

```python
 print('선형 회귀 모델 결과')
print('절편', lr.intercept_, '계수', lr.coef_)
```

```python
print(lr.score(x,y))
```

```python
x_new=[[19],[64]]
y_hat=lr.predict(x_new)
print(y_hat)
```

```python
plt.figure(figsize=(10,5))
plt.plot(x_new, y_hat, '-r')
plt.plot(x, y, 'b.')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

### 2. 정규방정식
(1) scikit-learn의 SGDRegressor  
* 코드구현

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import SGDRegressor

data= pd.read_csv('../data/insurance.csv')
x=np.array(data['age'])
y=np.array(data['charges'])
x=x.reshape(1338 ,1) 
# .reshape()으로 차원 맞추기
#y=y.reshape(1338 ,)

sgd_reg =SGDRegressor(max_iter=1000, random_state=34)
#sgd_reg.fit(x,y.ravel())
sgd_reg.fit(x,y)
```

```python
 print('SGD 회귀 모델 결과')
print('절편', sgd_reg.intercept_, '계수', sgd_reg.coef_)
```

```python
x_new=[[19],[64]]
y_hat=sgd_reg.predict(x_new)
print(y_hat)
```

```python
plt.figure(figsize=(10,5))
plt.plot(x_new, y_hat, '-r')
plt.plot(x, y, 'b.')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
