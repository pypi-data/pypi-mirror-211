### 명사만 추출

```python
# 단어로 토큰화

from nltk import word_tokenize
words = word_tokenize(line)
print(type(words), len(words))
print(words)
```

```python
# 품사 태깅 후 명사만 선택 / 형용사만 선택

nltk.tag.pos_tag(words)
```
